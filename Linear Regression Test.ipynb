```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
import pandas as pd
data = {"Size":[6, 8, 10 ,14, 16],
       "Price":[350, 775, 1150, 1395, 1675]
       }
data = pd.DataFrame(data)
```

```python
X = data["Size"].to_numpy().reshape(-1, 1)
y = data["Price"]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.show()
```

```python
lr.fit(X, y)
```

```python
import numpy as np
ls = []
for x in X:
    ls.append(lr.predict([x]))
```

```python
np.array(ls).reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, ls)
plt.show()
```
