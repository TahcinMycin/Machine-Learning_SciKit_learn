### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### CSV file

```python
data = pd.read_csv('data.csv')
```

```python
X = data["Size"]
y = data["Price"]
```

```python
X = X.to_numpy().reshape(-1, 1)
```

```python
y = y.to_numpy().reshape(-1, 1)
```

### Plotting data

```python
plt.scatter(X, y)
plt.xlabel("Pizza Size")
plt.ylabel("Pizza Price")
plt.title("Machine Learning on Pizza theory")
plt.style.use('ggplot')
plt.grid()
plt.show()
```

## Machine Learning Model

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

Let's predict some values

```python
model.predict([[25]])
```

```python
for i in X:
  pred = model.predict(X)
```

#### Plotting Best Fit Line

```python
plt.scatter(X, y)
plt.xlabel("Pizza Size")
plt.ylabel("Pizza Price")
plt.title("Machine Learning on Pizza theory")
plt.style.use('ggplot')
plt.plot(X, pred)
plt.grid()
plt.show()
```

#### Model's accuracy score

```python
model.score(X, y)
```
